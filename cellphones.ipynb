```python
def get_brand_info(url):
    browser = webdriver.Chrome()
    browser.get(url)
    
    sleep(randint(3,6))
    
    name_brands = []
    link_brands = []
    
    list_brands = browser.find_elements(By.CLASS_NAME, 'list-brand')
    
    for b in list_brands:
        tag_link = b.find_elements(By.CLASS_NAME, 'list-brand__item.button__link')
        for t in tag_link:
            #Get link brand
            link = t.get_attribute('href')
            
            link_segments = link.split("/")[-1]
            name_brand = link_segments.split('.')[0]
            
            print(link)
            print(name_brand)
            
            link_brands.append(link)
            name_brands.append(name_brand)
    
    browser.close()
    return name_brands, link_brands
```

```python
from concurrent.futures import ThreadPoolExecutor
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
from selenium import webdriver

def get_product_info(url):
    browser = webdriver.Chrome()
    browser.get(url)
    
    sleep(randint(3,6))
    
    wait = WebDriverWait(browser, 10)
    sleep(15)
    element = wait.until(ec.element_to_be_clickable((By.CLASS_NAME, 'cancel-button-top')))
    sleep(15)
    element.click()
    
    names = []
    links = []
    current_prices = []
    old_prices = []
    percent_discounts = []
    
    show_more_btn = browser.find_elements(By.CLASS_NAME, 'button.btn-show-more.button__show-more-product')
    if len(show_more_btn) > 0:
        while True:
            try:
                show_more_btn[0].click()
                sleep(randint(5,10))
            except Exception as e:
                break
        
    products = browser.find_elements(By.CLASS_NAME, 'product-info-container.product-item')
    
    # Define a function to extract product information
    def extract_product_info(product):
        name = product.find_elements(By.CLASS_NAME, 'product__name')
        if len(name) > 0:
            print(name[0].text)
            names.append(name[0].text)
        else:
            print('No name')
            names.append('No name')

        link = product.find_elements(By.CLASS_NAME, 'product__link.button__link')
        if len(link) > 0:
            print(link[0].get_attribute('href'))
            links.append(link[0].get_attribute('href'))
        else:
            print('No link found')
            links.append('No link found')

        current_price = product.find_elements(By.CLASS_NAME, 'product__price--show')
        if len(current_price) > 0:
            print(current_price[0].text)
            current_prices.append(current_price[0].text)
        else:
            print("Giá liên hệ")
            current_prices.append("Giá liên hệ")

        old_price = product.find_elements(By.CLASS_NAME, 'product__price--through')
        if len(old_price) > 0:
            print(old_price[0].text)
            old_prices.append(old_price[0].text)
        else:
            print("0")
            old_prices.append("0")

        percent_discount = product.find_elements(By.CLASS_NAME, 'product__price--percent')
        if len(percent_discount) > 0:
            print(percent_discount[0].text)
            percent_discounts.append(percent_discount[0].text)
        else:
            print("Giảm 0%")
            percent_discounts.append("Giảm 0%")
    
    # Use ThreadPoolExecutor to execute the extraction function for each product
    with ThreadPoolExecutor(max_workers=5) as executor:
        executor.map(extract_product_info, products)
    
    browser.close()
    return names, links, current_prices, old_prices, percent_discounts
```

```python
def build_brand_data_frame(name_brands, link_brands):
    
    brand = pd.DataFrame()
    
    brand['name'] = name_brands
    brand['link'] = link_brands
        
    return brand
```

```python
from itertools import product


def build_product_data_frame(names, links, current_prices, old_prices, percent_discounts):
    product = pd.DataFrame()
    
    product['product_name'] = names
    product['prodcut_link'] = links
    product['current_prices'] = current_prices
    product['old_prices'] = old_prices
    product['percent_discounts'] = percent_discounts
    
    return product
```

```python
name_brands, link_brands = \
get_brand_info('https://cellphones.com.vn/mobile.html')
```

```python
names, links, current_prices, old_prices, percent_discounts = \
get_product_info('https://cellphones.com.vn/mobile.html')
```
